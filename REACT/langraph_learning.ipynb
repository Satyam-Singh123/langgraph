```python
from langgraph.graph import StateGraph, MessagesState, START, END

def mock_llm(state: MessagesState):
    return {"messages": [{"role": "ai", "content": "hello world"}]}


# Create the graph structure based on the MessagesState schema
graph = StateGraph(MessagesState)

# (Corrected Syntax for clarity)
graph.add_node("mock_llm", mock_llm)

# Define the flow (Edges)
graph.add_edge(START, "mock_llm")
graph.add_edge("mock_llm", END)

# Finalize the graph structure and make it runnable
#Turns the conceptual graph structure into a fixed, optimized, and runnable object (called a Runnable)
graph = graph.compile()

#invoke(): Executes the compiled graph once.
# Input: The input is a dictionary that matches the MessagesState schema.
graph.invoke({"messages": [{"role": "user", "content": "hi!"}]})
```
